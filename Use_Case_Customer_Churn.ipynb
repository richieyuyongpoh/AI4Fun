### Initialization

1: Read the csv and save as a dataframe/table

```python

import pandas as pd

data = pd.read_table('https://raw.githubusercontent.com/richieyuyongpoh/Practical_Python_series/main/churn.txt', index_col=False,  sep=',',skipinitialspace=True)

data
```

2: As the `phnum` is the `primary key` or `id` for this table/dataset, remove it. 

```python
data = data.drop('phnum',axis=1)

data.head()
```

3: Get the simple descriptive statistics

```python

data.info()
data.describe()
```

4: Define the Features , `X`

```python
X = data.drop('label',axis=1)


X.head()
```

5: Remove the categorical variables that are written in texts, such as `st`, `intplan`, `voice`.
<br><br>
Note: in the later part, we will pre-process these categorical variables and then include them for the modelling. Refer to step [#14](https://colab.research.google.com/drive/1ndH11fsmbKMX7S-xtRMqAOCBcXnwxLOs?authuser=1#scrollTo=2zdN6QPfdzON&line=3&uniqifier=1)

```python
X = X.drop(['st','intplan','voice'],axis=1)


X.head()
```

6: define the label, `y`

```python
y = data['label']
y.head()
```

7: For the quick start, let us ignore pre-processing steps and directly move on to the machine learning (training & testing) steps.

Let us start with `train-test-split()`

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)
```

```python
Xtrain.head()
```

```python
ytrain.head()
```

### *k*-nearest neigbours (*k*-NN model)

8: Apply *k*-NN model. 

```python

from sklearn.neighbors import KNeighborsClassifier


knn = KNeighborsClassifier()
knn.fit(Xtrain, ytrain)
ypred = knn.predict(Xtest)


```

9: Observe the result

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(ytest, ypred))
print()
print()
print(classification_report(ytest, ypred))
```

### Support Vector Machine (SVM)

10: Apply SVM and observe the results

```python
from sklearn.svm import SVC

svc = SVC()
svc.fit(Xtrain, ytrain)
ypred = svc.predict(Xtest)

print(confusion_matrix(ytest, ypred))
print()
print()
print(classification_report(ytest, ypred))
```

### Logistic Regression

11: Apply Logistic Regression and observe the results

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(Xtrain, ytrain)
ypred = logreg.predict(Xtest)

print(confusion_matrix(ytest, ypred))
print()
print()
print(classification_report(ytest, ypred))
```

### Gaussian Naive Bayes

12: Apply NB and observe the results

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(Xtrain, ytrain)
ypred = nb.predict(Xtest)

print(confusion_matrix(ytest, ypred))
print()
print()
print(classification_report(ytest, ypred))
```

### Random Forest

13: Apply RF and observe the results

```python
from sklearn.ensemble import RandomForestClassifier

RandomForest = RandomForestClassifier()
RandomForest.fit(Xtrain, ytrain)
ypred = RandomForest.predict(Xtest)

print(confusion_matrix(ytest, ypred))
print()
print()
print(classification_report(ytest, ypred))
```

It seems Random Forest gives the best results.

### Improvements

14: Apply label encoder to the categorical variables

Refer to step [#5](https://colab.research.google.com/drive/1ndH11fsmbKMX7S-xtRMqAOCBcXnwxLOs?authuser=1#scrollTo=JLHTAof1Ps6f&line=3&uniqifier=1)

```python
from sklearn.preprocessing import LabelEncoder

data = pd.read_table('https://raw.githubusercontent.com/richieyuyongpoh/Practical_Python_series/main/churn.txt', index_col=False,  sep=',',skipinitialspace=True)
data = data.drop('phnum',axis=1)

labelencoder = LabelEncoder()
data['st'] = labelencoder.fit_transform(data['st'])
data['intplan'] = labelencoder.fit_transform(data['intplan'])
data['voice'] = labelencoder.fit_transform(data['voice'])

data.head()
```

15: Repeat #4, #6, and #7

```python
X = data.drop('label',axis=1)

y = data['label']

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y)
```

16: Repeat #13 

```python
from sklearn.ensemble import RandomForestClassifier

RandomForest = RandomForestClassifier()
RandomForest.fit(Xtrain, ytrain)
ypred = RandomForest.predict(Xtest)

print(confusion_matrix(ytest, ypred))
print()
print()
print(classification_report(ytest, ypred))
```

### What are the leading factors that affect churn rate?

17: check the feature importance

```python
RandomForest.feature_importances_
```

```python
list(X.columns)
```

```python
important_factors = pd.DataFrame({'Factor': list(X.columns), 'Importance': RandomForest.feature_importances_})
important_factors
```

```python
important_factors.sort_values(by=['Importance'], ascending=False,inplace=True)
important_factors
```

Random Forest gives the best result. By doing some enhancement steps, the accuracy increases to 97%++ while recall (for class 1 - churn ) increases to 80%++ .
